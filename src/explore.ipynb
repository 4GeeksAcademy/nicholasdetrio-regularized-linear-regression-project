```python
# Making 3 new columns that are the numeric representation of the original categorical column
total_data['COUNTY_NAME_n'] = pd.factorize(total_data['COUNTY_NAME'])[0]
total_data['STATE_NAME_n'] = pd.factorize(total_data['STATE_NAME'])[0]
#total_data['room_type_n'] = pd.factorize(total_data['room_type'])[0]

total_data.head()
```

```python
# Drop the original categorical columns
total_data = total_data.drop(['COUNTY_NAME', 'STATE_NAME'], axis=1)
total_data.head()
```

```python
# X represents all features excluding the target
X = total_data # A dataframe (double brackets)
X = total_data.drop(columns=['Heart disease_number'])

# y contains our target feature
y = total_data['Heart disease_number'] # A series (single brackets)
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
nrows, ncolumns = total_data.shape

print(f"Our entire dataset has {nrows} rows.")
print(f"Our training set should have 80% of our data so it should have {int(0.8*nrows)} rows.")
print(f"Our test set should have 20% of our data so it should have {int(0.2*nrows)} rows.")
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
print(f"Intercep (a): {model.intercept_}")
print(f"Coefficients: {model.coef_}")
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print(f"MSE: {mean_squared_error(y_test, y_pred)}")
print(f"R2 Score: {r2_score(y_test, y_pred)}")
```

```python
from sklearn.linear_model import Lasso

alpha = 1.0
lasso_model = Lasso(alpha = alpha)

# Entrenamos el modelo
lasso_model.fit(X_train, y_train)

# Evaluamos el rendimiento del modelo en los datos de prueba
score = lasso_model.score(X_test, y_test)
print("Coefficients:", lasso_model.coef_)
print("R2 score:", score)
```

```python
from pickle import dump
# from pickle import dump is a Python statement that imports the dump function from the pickle module.
# pickle is a module in Python that provides functions for serializing and deserializing Python objects. The dump function is used to serialize (i.e., convert into a byte stream) a Python object and write it to a file.

dump(lasso_model, open("../models/lasso_alpha-1.0.sav", "wb"))
```
